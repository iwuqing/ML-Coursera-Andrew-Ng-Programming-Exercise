# Part 3: Cost and Gradient descent

## 3-1 The implementations of **Cost funtion** and **Gradient Descent**

```python
# cost funtion
def compute_cost(X, Y, theta):
    m = len(Y) # number of training example
    J = np.dot((np.dot(X, theta)-Y).T, (np.dot(X, theta)-Y)) / (2*m)
    return J

# gradient descent
def gradient_descent(X, Y, theta, iters, alpha):
    m = len(Y) # number of training example
    J_history = np.zeros([iters, 1])
    
    for i in range(0, iters):
        theta = theta - alpha/m * np.dot(X.T, (np.dot(X, theta)-Y))
        J_history[i, 0] = compute_cost(X, Y, theta)
    return theta, J_history
```

## 3-2 Some settings for Gradient Descent and Cost Funtion

```python
# some gradient descent settings
iterations = 1500;
alpha = 0.01;
X = np.hstack((np.ones([m, 1], dtype=float), X)) # add a column of ones to X
theta = np.zeros([2, 1]); #initialize fitting parameters
```

## 3-3 Test the cost function

```python
# compute and display initial cost
J = compute_cost(X, Y, theta)
print("With theta = [0; 0]\nCost computed = " + str(J))
print("Expected cost value(approx) 32.07")

# further testing of the cost funtion
J = compute_cost(X, Y, np.array([[-1],[2]]))
print("With theta = [-1; 2]\nCost computed = " + str(J))
print("Expected cost value(approx) 54.24")
```

## 3-4 Run Gradient descent

```python
theta, _ = gradient_descent(X, Y, theta, iterations, alpha)
# print theta to sceen
print("Theta found by gradient descent:")
print(theta)
print("Expected theta value(approx)\n-3.6303  1.1664")
```

## 3-5 Plot the linear fit

```python
plt.plot(X[:, 1], np.dot(X, theta), "r-", label="linear regression")
plt.legend()
plt.scatter(X[:, 1], Y, label="training data")
plt.legend()
plt.xlabel("Profit in $10,100s")
plt.ylabel("Population of City in 10,100s")
plt.show()
```

## 3-6 Predict values for population size of 35,000 and 70,000

```python
preditc1 = np.dot(np.array([[1, 3.5]]), theta)
print("for population = 35,000, we predict a profit of " + str(10000*preditc1))

preditc2 = np.dot(np.array([[1, 7]]), theta)
print("for population = 70,000, we predict a profit of " + str(10000*preditc2))
```

# Part 4: Visualizing J($\theta_0, \theta_1$)

```python
theta0_value = np.linspace(-10, 10, 100)
theta1_value = np.linspace(-1, 4, 100)

J_value = np.zeros([len(theta0_value), len(theta1_value)])

# fill out J_value
for i in range(0, len(theta0_value)):
    for j in range(0, len(theta1_value)):
        t = np.array([[theta0_value[i]], [theta1_value[j]]])
        J_value[i, j] = compute_cost(X, Y, t)
```

```python
# surface plot
from mpl_toolkits.mplot3d import Axes3D

ax = Axes3D(plt.figure())
ax.plot_surface(theta1_value, theta0_value, J_value.T, cmap="rainbow")
ax.set_xlabel("theta1")
ax.set_ylabel("theta0")
ax.set_zlabel("J")
```

```python
# contour plot
plt.contourf(theta0_value, theta1_value, J_value, 20)
plt.contour(theta0_value, theta1_value, J_value, colors="black")
plt.show()
```
